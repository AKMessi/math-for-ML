## 3. NumPy Implementation From Scratch

Use the local teaching module or plain NumPy to implement the core idea explicitly.

````python
from src.distributions import gaussian_pdf, maximum_likelihood_gaussian

samples = np.array([1.2, 0.9, 1.4, 1.1, 1.0])
mle_mean, mle_std = maximum_likelihood_gaussian(samples)
x = np.linspace(-1, 3, 200)
numpy_result = gaussian_pdf(x, mle_mean, mle_std)
print(mle_mean, mle_std)
````

## 4. PyTorch Verification

Rebuild the same quantity in PyTorch and verify equivalence with `torch.allclose`.

````python
dist = torch.distributions.Normal(torch.tensor(mle_mean), torch.tensor(mle_std))
torch_result = torch.exp(dist.log_prob(torch.tensor(x, dtype=torch.float64)))
print(torch.allclose(torch_result, torch.tensor(numpy_result), atol=1e-5))
````

## 5. Visualization

Connect the formula to geometry, distributions, optimization behavior, or spectral structure.

````python
plt.hist(samples, bins=5, density=True, alpha=0.4)
plt.plot(x, numpy_result)
plt.title('Estimated Gaussian density')
plt.show()
````
