## 3. NumPy Implementation From Scratch

Use the local teaching module or plain NumPy to implement the core idea explicitly.

````python
from src.optimizers import Adam, minimize

objective = lambda w: float(np.sum((w - 3.0) ** 2))
gradient = lambda w: 2.0 * (w - 3.0)
result = minimize(objective, gradient, np.array([8.0]), Adam(learning_rate=0.2), iterations=25)
numpy_result = result.parameters
print(numpy_result)
````

## 4. PyTorch Verification

Rebuild the same quantity in PyTorch and verify equivalence with `torch.allclose`.

````python
parameter = torch.tensor([8.0], dtype=torch.float64, requires_grad=True)
optimizer = torch.optim.Adam([parameter], lr=0.2)
for _ in range(25):
    optimizer.zero_grad()
    loss = torch.sum((parameter - 3.0) ** 2)
    loss.backward()
    optimizer.step()
torch_result = parameter.detach()
print(torch.allclose(torch_result, torch.tensor(numpy_result), atol=5e-2))
````
