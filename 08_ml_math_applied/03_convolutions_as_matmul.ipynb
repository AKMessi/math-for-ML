## 3. NumPy Implementation From Scratch

Use the local teaching module or plain NumPy to implement the core idea explicitly.

````python
from src.ml_components import scaled_dot_product_attention

Q = np.array([[[1.0, 0.0], [0.0, 1.0]]])
K = np.array([[[1.0, 0.0], [0.0, 1.0]]])
V = np.array([[[2.0, 1.0], [0.5, 3.0]]])
output, weights = scaled_dot_product_attention(Q, K, V)
numpy_result = output
print(weights)
````

## 4. PyTorch Verification

Rebuild the same quantity in PyTorch and verify equivalence with `torch.allclose`.

````python
Q_t = torch.tensor(Q, dtype=torch.float64)
K_t = torch.tensor(K, dtype=torch.float64)
V_t = torch.tensor(V, dtype=torch.float64)
scores = Q_t @ K_t.transpose(-1, -2) / np.sqrt(Q.shape[-1])
torch_result = torch.softmax(scores, dim=-1) @ V_t
print(torch.allclose(torch_result, torch.tensor(numpy_result), atol=1e-5))
````
