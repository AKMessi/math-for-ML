## 3. NumPy Implementation From Scratch

Use the local teaching module or plain NumPy to implement the core idea explicitly.

````python
from src.linear_algebra import pca_from_scratch

X = np.array([[2., 1.], [3., 2.], [4., 3.], [5., 3.5]])
result = pca_from_scratch(X, n_components=1)
numpy_result = result.scores
print(result.components)
print(result.explained_variance_ratio)
````

## 4. PyTorch Verification

Rebuild the same quantity in PyTorch and verify equivalence with `torch.allclose`.

````python
X_t = torch.tensor(X, dtype=torch.float64)
centered = X_t - X_t.mean(dim=0)
_, _, Vh = torch.linalg.svd(centered, full_matrices=False)
torch_result = centered @ Vh[:1].T
print(torch.allclose(torch_result.abs(), torch.tensor(numpy_result).abs(), atol=1e-5))
````
