## 3. NumPy Implementation From Scratch

Use the local teaching module or plain NumPy to implement the core idea explicitly.

````python
from src.autodiff import Value, gradient_check

x = Value(1.5)
y = (x * x + 3.0 * x + 1.0).tanh()
y.backward()
numpy_result = np.array([x.grad])
print('autodiff gradient =', x.grad)
print(gradient_check(lambda nodes: (nodes[0] * nodes[0] + 3.0 * nodes[0] + 1.0).tanh(), [1.5]))
````

## 4. PyTorch Verification

Rebuild the same quantity in PyTorch and verify equivalence with `torch.allclose`.

````python
x_t = torch.tensor(1.5, dtype=torch.float64, requires_grad=True)
y_t = torch.tanh(x_t * x_t + 3.0 * x_t + 1.0)
y_t.backward()
torch_result = torch.tensor([x_t.grad], dtype=torch.float64)
print(torch.allclose(torch_result, torch.tensor(numpy_result), atol=1e-5))
````

## 5. Visualization

Connect the formula to geometry, distributions, optimization behavior, or spectral structure.

````python
grid = np.linspace(-3, 3, 200)
curve = np.tanh(grid**2 + 3.0 * grid + 1.0)
plt.plot(grid, curve)
plt.title('A differentiable objective')
plt.show()
````
