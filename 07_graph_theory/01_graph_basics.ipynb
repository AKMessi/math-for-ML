## 3. NumPy Implementation From Scratch

Use the local teaching module or plain NumPy to implement the core idea explicitly.

````python
from src.graphs import Graph

graph = Graph()
graph.add_edge('A', 'B', weight=1.0)
graph.add_edge('B', 'C', weight=2.0)
graph.add_edge('A', 'C', weight=4.0)
adjacency = graph.adjacency_matrix()
numpy_result = adjacency
print(graph.bfs('A'))
print(graph.shortest_path('A', 'C'))
````

## 4. PyTorch Verification

Rebuild the same quantity in PyTorch and verify equivalence with `torch.allclose`.

````python
torch_result = torch.tensor(adjacency, dtype=torch.float64)
print(torch.allclose(torch_result, torch.tensor(numpy_result), atol=1e-5))
````
